## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
import seaborn as sns
data = df[['grade','price']]
sns.boxplot(x = 'grade', y = 'price', data = data)
plt.show()
#Hint: plt.boxplot
```

We are able to read much more information on the boxplot chart. This is important because in the chart above we see the median and percentiles, while in chart 2b we could incorrectly estimate the median, especially for the middle grade values, because in chart 2b the data on apartments merge and it is not clear whether there are several apartments in a given place or several hundred

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:
mosaic(df,index = ['bedrooms','grade'])
plt.show()
```

This mosaic plot illustrates the relationship between the number of bedrooms and the quality of properties. It shows how different combinations of bedroom counts and quality levels are distributed in the dataset. The plot helps identify patterns and potential associations between these variables.
