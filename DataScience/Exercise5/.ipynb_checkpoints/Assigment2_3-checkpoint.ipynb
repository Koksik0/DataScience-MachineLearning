#### A correlation matrix is a table that illustrates the interdependence between variables. In statistical analysis, this matrix contains correlation coefficients between all parameters in a given sample. Correlation values between -1 and 1 provide information about the strength and direction of the relationship between variables. Positive values indicate positive correlation, negative values indicate negative correlation, and 0 indicates no correlation. The closer the absolute value is to 1, the stronger the relationship. However, it's important to note that correlation does not imply causation; there may be a relationship, but it doesn't necessarily mean that one variable causes changes in the other.

# Regression models

```python
from sklearn.linear_model import LinearRegression
X = df.age.values.reshape(-1,1)[:1000]
y = df.smoking.values.reshape(-1)[:1000]

model = LinearRegression() 
model.fit(X, y)
```

```python
y_pred = model.predict(X)
plt.figure(figsize=(12,5), dpi=80)
plt.scatter(X, y, color = 'blue')
plt.plot(X, y_pred, color = 'red')
plt.show()
```

#### After creating a linear regression model, it is evident that the older the person, the less likely he or she smokes

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
fig, axes = plt.subplots(4, 2, figsize=(12, 16))
mean_squared_error_list = []
for order, ax in zip(range(1, 9), axes.ravel()):
    poly = PolynomialFeatures(degree=order)
    X_order = poly.fit_transform(X)
    
    temp = LinearRegression()
    temp.fit(X_order, y)
    
    sorted_X, sorted_y_pred = zip(*sorted(zip(X, temp.predict(X_order))))
    
    ax.scatter(X, y, color='blue')
    ax.plot(sorted_X, sorted_y_pred, color='red')
    ax.set_title(f'Degree {order}')
    
    y_pred = temp.predict(X_order)
    mse = mean_squared_error(y, y_pred)
    mean_squared_error_list.append(mse)
    
plt.tight_layout()
plt.show()
```

#### After creating 8 different polynomial regression models with degrees from 1 to 8, we can see that from the 4th degree the curve begins to fit the data more and more

```python
plt.figure(figsize=(14, 8))
plt.plot([x for x in range(1, 9)], mean_squared_error_list, linestyle='-', marker='o')
plt.xlabel('Degree of Polynomial')
plt.ylabel('MSE')
plt.title('Performance')
plt.grid(True)
plt.show()
```

#### In the graphs above we see polynomial curves of varying degrees. As you can easily see, the higher the level of the polynomial, the better the fit, as can be seen in the last graph comparing the effectiveness of these polynomials. However, for a polynomial of degree 8, we see that the better fit was for the 7th degree.

```python
mean_squared_error_list_test = []
mean_squared_error_list_train = []
for order in range(1,9):
    poly = PolynomialFeatures(degree=order)
    X_order = poly.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(X_order, y, test_size=0.3, random_state=44)
    
    temp = LinearRegression()
    temp.fit(X_train, y_train)
    
    y_pred = temp.predict(X_train)
    mse = mean_squared_error(y_train, y_pred)
    mean_squared_error_list_train.append(mse)
    
    y_pred1 = temp.predict(X_test)
    mse1 = mean_squared_error(y_test, y_pred1)
    mean_squared_error_list_test.append(mse1)

plt.figure(figsize=(14, 8))
plt.plot([x for x in range(1, 9)], mean_squared_error_list_train, linestyle='-', marker='o', color = 'blue')
plt.plot([x for x in range(1, 9)], mean_squared_error_list_test, linestyle='-', marker='o', color = 'red')
plt.xlabel('Degree of Polynomial')
plt.ylabel('Mean squared error')
plt.title('Performance')
plt.grid(True)
plt.show()
```

#### As you can see in this chart, the higher the degree of the polynomial, the more and more the test data differs from the training data. This may be due to overfitting as higher polynomials adapt better to the data and therefore to noise and irregularities.

```python

X = df[['age', 'height(cm)', 'weight(kg)', 'eyesight(left)', 'eyesight(right)', 'Cholesterol']].values.reshape(-1,6)
y = df.smoking.values.reshape(-1)

poly = PolynomialFeatures(3, include_bias = False)
X = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=43)
    
model = LinearRegression() 
_= model.fit(X_train, y_train) 

y_pred = model.predict(X_test)
print("Mean squared error equals: {0}".format(mean_squared_error(y_pred, y_test)))
```
